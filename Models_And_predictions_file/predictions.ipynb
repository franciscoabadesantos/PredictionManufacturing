```python
import pandas as pd
import os
print(os.getcwd())  # Print the current working directory

csv_file_name = "dataset_station4.csv"
data=pd.read_csv(csv_file_name)
data
```

```python
data['skill']=data['skill'].replace("Glue Type A",'1')
data['skill']=data['skill'].replace("Glue Type B",'2')
data['skill']=data['skill'].replace("Glue Type C",'3')

data
```

```python
from sklearn.model_selection import train_test_split

# Splitting the data based on the skill column
X = data[['velocity', 'energy_consumption']]
y = data['skill'].to_frame()

# Splitting into training and testing sets based on skill for velocity
X_velocity_train, X_velocity_test, y_velocity_train, y_velocity_test = train_test_split(y,
    X['velocity'], test_size=0.2, random_state=42)

# Splitting into training and testing sets based on skill for energy
X_energy_train, X_energy_test, y_energy_train, y_energy_test = train_test_split(y,
    X['energy_consumption'], test_size=0.2, random_state=42)
```

```python
X_energy_test
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, BayesianRidge, SGDRegressor
from sklearn.metrics import mean_squared_error, r2_score


# Define the models to test
models = [LinearRegression(), Ridge(), Lasso(), ElasticNet(),BayesianRidge(), SGDRegressor(), BayesianRidge()]

# Initialize variables to track the best model and its performance
best_velocity_model = None
best_velocity_mse = float('inf')
best_velocity_r2 = -float('inf')
best_energy_model = None
best_energy_mse = float('inf')
best_energy_r2 = -float('inf')



# Iterate over each velocity model
for model in models:
    # Train the velocity model
    model.fit(X_velocity_train, y_velocity_train)

    # Make predictions for velocity on the testing set
    velocity_predictions = model.predict(X_velocity_test)

    # Calculate MSE and R2 score for velocity predictions
    velocity_mse = mean_squared_error(y_velocity_test, velocity_predictions)
    velocity_r2 = r2_score(y_velocity_test, velocity_predictions)

    # Print the performance metrics for velocity prediction
    print("Velocity Model:", type(model).__name__)
    print("Velocity MSE:", velocity_mse)
    print("Velocity R2 score:", velocity_r2)
    print()

    # Update the best velocity model if the current model performs better
    if velocity_mse < best_velocity_mse or (velocity_mse == best_velocity_mse and velocity_r2 > best_velocity_r2):
        best_velocity_model = model
        best_velocity_mse = velocity_mse
        best_velocity_r2 = velocity_r2
        
        
# Iterate over each energy model
for model in models:
    # Train the energy model
    model.fit(X_energy_train, y_energy_train)

    # Make predictions for energy on the testing set
    energy_predictions = model.predict(X_energy_test)

    # Calculate MSE and R2 score for energy predictions
    energy_mse = mean_squared_error(y_energy_test, energy_predictions)
    energy_r2 = r2_score(y_energy_test, energy_predictions)

    # Print the performance metrics for energy prediction
    print("Energy Model:", type(model).__name__)
    print("Energy MSE:", energy_mse)
    print("Energy R2 score:", energy_r2)
    print()

    # Update the best energy model if the current model performs better
    if energy_mse < best_energy_mse or (energy_mse == best_energy_mse and energy_r2 > best_energy_r2):
        best_energy_model = model
        best_energy_mse = energy_mse
        best_energy_r2 = energy_r2


print("Best Velocity Model:", type(best_velocity_model).__name__)
print("Best Velocity MSE:", best_velocity_mse)
print("Best Velocity R2 score:", best_velocity_r2)

print("Best Energy Model:", type(best_energy_model).__name__)
print("Best Energy MSE:", best_energy_mse)
print("Best Energy R2 score:", best_energy_r2)
```

```python
import numpy as np

# Extract the features (velocity and energy_consumption)
new_data = np.array([[1]])

# Make predictions using the best velocity model
best_velocity_model.fit(X_velocity_train, y_velocity_train)
velocity_predicted_values = best_velocity_model.predict(new_data)

# Make predictions using the best energy model
best_energy_model.fit(X_energy_train, y_energy_train)
energy_predicted_values = best_energy_model.predict(new_data)

# Print the predicted values
predicted_velocity = velocity_predicted_values[0]  # Assuming the first value corresponds to velocity
predicted_energy = energy_predicted_values[0]  # Assuming the first value corresponds to energy

print("Predicted velocity:", predicted_velocity)
print("Predicted energy consumption:", predicted_energy)
```

```python
import joblib

# Save the best_velocity_model
file_name_velocity = "best_velocity_model_" + csv_file_name + "_" + type(best_velocity_model).__name__ + ".joblib"
joblib.dump(best_velocity_model, file_name_velocity)

# Save the best_energy_model
file_name_energy = "best_energy_model_" + csv_file_name + "_" + type(best_energy_model).__name__ + ".joblib"
joblib.dump(best_energy_model, file_name_energy)
```
